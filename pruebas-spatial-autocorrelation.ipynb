```python
import pandas as pd
import pyarrow as pa
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import geopandas as gpd
from libpysal import weights
from esda.moran import Moran_Local, Moran_Local_BV
from splot.esda import moran_scatterplot, lisa_cluster
from spreg import OLS, GM_Lag, GM_Error_Het
```

# Procesamiento de datos

## Índice de Gini

```python
d_gini = pd.read_csv("/kaggle/input/gini-ine/gini.csv", sep=";", decimal=",")
d_gini = d_gini[(d_gini["Periodo"] == 2020) &
                (d_gini["Indicadores de renta media"] == "Índice de Gini") &
                (d_gini["Distritos"].isna()) &
                (d_gini["Secciones"].isna())]
d_gini["Municipios_Code"] = d_gini["Municipios"].str[0:5]
d_gini["Municipios_Name"] = d_gini["Municipios"].str[6:]
d_gini['Gini'] = pd.to_numeric(d_gini['Total'].str.replace(",", "."), errors="coerce")
d_gini = d_gini[["Municipios_Code", "Municipios_Name", "Periodo", "Gini"]] \
    .set_index("Municipios_Code") \
    .dropna()
d_gini.head()
```

## Datos demográficos

```python
d_demo = pd.read_csv("/kaggle/input/demograficos-ine/demograficos.csv", sep=";", decimal=",", encoding='latin-1')
d_demo["Municipios_Code"] = d_demo["Municipio"].str[0:5]
d_demo = d_demo[["Porcentaje de población menor de 18 años", 
                 "Porcentaje de población de 65 y más años", 
                 "Porcentaje de hogares unipersonales",
                 "Municipios_Code"]] \
    .set_index("Municipios_Code")
for col in d_demo.columns:
    d_demo[col] = d_demo[col].str.replace("\.", "")
    d_demo[col] = pd.to_numeric(d_demo[col].str.replace(",", "."), errors="coerce")
d_demo = d_demo.dropna()
d_demo.head()
```

## Datos de empleo por cuenta ajena

```python
d_ajena = pd.read_csv("/kaggle/input/cuenta-ajena/cuenta_ajena.csv", sep=";", decimal=",")
d_ajena = d_ajena[(d_ajena["Municipios"] != "Total Nacional") &
                  (d_ajena["Indicadores"] == "Tasa Neta de Crecimiento") & 
                  (d_ajena["Periodo"] == "2020T1")]
d_ajena["Municipios_Code"] = d_ajena["Municipios"].str[0:5]
d_ajena["Municipios_Name"] = d_ajena["Municipios"].str[6:]
d_ajena['Crecimiento Empleo Cuenta Ajena'] = pd.to_numeric(d_ajena['Total'].str.replace(",", "."), errors="coerce")
d_ajena = d_ajena[["Municipios_Code", "Crecimiento Empleo Cuenta Ajena"]] \
    .set_index("Municipios_Code") \
    .dropna()
d_ajena.head()
```

## Datos de empleo por cuenta propia

```python
d_propia = pd.read_csv("/kaggle/input/cuenta-propia/cuenta_propia.csv", sep=";", decimal=",")
d_propia = d_propia[(d_propia["Municipios"] != "Total Nacional") &
                  (d_propia["Indicadores"] == "Tasa Neta de Crecimiento") & 
                  (d_propia["Periodo"] == "2020T1")]
d_propia["Municipios_Code"] = d_propia["Municipios"].str[0:5]
d_propia["Municipios_Name"] = d_propia["Municipios"].str[6:]
d_propia['Crecimiento Empleo Cuenta Propia'] = pd.to_numeric(d_propia['Total'].str.replace(",", "."), errors="coerce")
d_propia = d_propia[["Municipios_Code", "Crecimiento Empleo Cuenta Propia"]] \
    .set_index("Municipios_Code") \
    .dropna()
d_propia.head()
```

## Datos geográficos

```python
# http://centrodedescargas.cnig.es/CentroDescargas/index.jsp
gd = pd.concat([
    gpd.read_file("/kaggle/input/poly-peninbal/poly_peninbal.shp").to_crs("epsg:4326")
    # gpd.read_file("/kaggle/input/poly-canarias/poly_canarias.shp").to_crs("epsg:4326")
])
gd = gd[["NATCODE", "geometry"]]
gd["NATCODE"] = gd["NATCODE"].str[-5:]
gd = gd \
        .rename(columns={"NATCODE": "Municipios_Code"}) \
        .set_index("Municipios_Code")
gd.head()
```

```python
d = d_gini \
    .join(d_demo, how="inner") \
    .join(d_ajena, how="inner") \
    .join(d_propia, how="inner") \
    .join(gd, how="inner")
d = d.drop("Periodo", axis=1)
d = gpd.GeoDataFrame(d)
d.head()
```

```python
d_plot = d.copy()
# Douglas-Peucker Algorithm to simplify geometries
d_plot.geometry = d_plot.geometry.simplify(0.005)
```

```python
def generate_map(col):
    return d_plot.explore(column=col, 
                          cmap="BuPu",
                          #scheme='HeadTailBreaks',
                          tooltip=False,
                          popup=True) 
widgets.interact(generate_map, 
                 col = widgets.Dropdown(options=d.drop(["Municipios_Name", "geometry"], axis=1).columns))
```

# Autocorrelación espacial (Índice de Moran)

```python
# Rook neighbors are those states that share an edge on their respective borders
# Queen neighbors are based on a more inclusive condition that requires only a shared vertex between two states
w = weights.contiguity.Queen.from_dataframe(d)
```

```python
y1 = d['Gini'].values
# https://nbviewer.org/github/pysal/splot/blob/main/notebooks/esda_morans_viz.ipynb
moran_loc = Moran_Local(y1, w)
```

```python
fig, ax = moran_scatterplot(moran_loc, p=0.05)
plt.show()
```

```python
lisa_cluster(moran_loc, d, p=0.05, figsize=(12,9))
plt.show()
```

```python
# Bivariate Moran
# Measure the degree one polygon with a specific attribute is correlated with its neighboring polygons with a different attribute
y2 = d['Crecimiento Empleo Cuenta Ajena'].values
moran_loc_bv = Moran_Local_BV(y1, y2, w)
fig, ax = moran_scatterplot(moran_loc_bv, p=0.05)
ax.set_xlabel('Gini') # y2
ax.set_ylabel('Spatial lag de Crecimiento Empleo Cuenta Ajena') # y1
plt.show()
```

```python
lisa_cluster(moran_loc_bv, d, p=0.05, figsize=(12,9))
plt.show()
```

# Regresión no espacial con modelo OLS (Ordinary Least Squares)
In a non-spatial OLS model, y=Xβ+ε, we examine the relationship between one or more independent variables and a dependent variable.

https://chrismgentry.github.io/Spatial-Regression/#12_Basic_Spatial_Regression_Models

https://geographicdata.science/book/notebooks/11_regression.html#exogenous-effects-the-slx-model

https://pysal.org/spreg/api.html

```python
d.columns
```

```python
model_1 = OLS(
    d[["Gini"]].values,  # Dependent variable
    d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).values,  # Independent variables
    name_y="Gini",  # Dependent variable name
    name_x=d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).columns.tolist()  # Independent variables names
)
```

```python
pd.DataFrame(
    {
        "Coeff": model_1.betas.flatten(),
        "StdError": model_1.std_err.flatten(),
        "PValue": [i[1] for i in model_1.t_stat]
    },
    index=model_1.name_x
).round(4).query('PValue < 0.05')
```

```python
print(model_1.summary)
```

# Regresión espacial con modelo SLX (Spatially Lagged X variables)
With a spatially lagged, or SLX model, we see an addition of a spatial component to the equation that accounts for the average value of neighboring X values. So the SLX equation examines whether variables that impact our neighbors also impact us. This is consdiered a local model as there is only one-way interaction between our neighbors and us.

```python
knn = weights.distance.KNN.from_dataframe(d.reset_index(), k=10)
knn.transform = "R"  # Row standardize weights
d_lagged = (
    d
    .drop(["Municipios_Name", "geometry"], axis=1)
    .apply(lambda y: weights.spatial_lag.lag_spatial(knn, y))
    .rename(columns=lambda c: "lag_" + c)
)
d_lagged = d.join(d_lagged).drop(["Municipios_Name", "geometry"], axis=1)
d_lagged.head()
```

```python
model_2 = OLS(
    d_lagged[["Gini"]].values,  # Dependent variable
    d_lagged.drop(["Gini", "lag_Gini"], axis=1).values,  # Independent variables
    name_y="Gini",  # Dependent variable name
    name_x=d_lagged.drop(["Gini", "lag_Gini"], axis=1).columns.tolist()  # Independent variables names
)
```

```python
pd.DataFrame(
    {
        "Coeff": model_2.betas.flatten(),
        "StdError": model_2.std_err.flatten(),
        "PValue": [i[1] for i in model_2.t_stat]
    },
    index=model_2.name_x
).round(4).query('PValue < 0.05')
```

```python
print(model_2.summary)
```

# Regresión espacial con modelo SLM (Spatially Lagged y variable)
In the Spatial Lag model, we lose the lagged X values that were used in the SLX model and instead add a spatially lagged y value. In this global model the dependent variable among our neighbors influences our dependent variable. Therefore there is a feedback loop that occurs where affects on our neighbor(s) y affects our y and our neighbor(s) y variable.

```python
model_3 = GM_Lag(
    d[["Gini"]].values,  # Dependent variable
    d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).values,  # Independent variables
    w=knn,
    name_y="Gini",  # Dependent variable name
    name_x=d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).columns.tolist()  # Independent variables names
)
```

```python
print(model_3.summary)
```

# Regresión espacial con modelo SEM (Spatial Error Model)
The Spatial Error model does not include lagged dependent or independent variables, but instead includes a function of our unexplained error and that of our neighbors. This is a global model where the higher than expected residual values suggest a missing explanatory variable that is spatially correlated. This would lead to unexplained clusters of spatially correlated values within our neighborhood that were not included as a predictor variable(s).

```python
model_4 = GM_Error_Het(
    d[["Gini"]].values,  # Dependent variable
    d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).values,  # Independent variables
    w=knn,
    name_y="Gini",  # Dependent variable name
    name_x=d.drop(["Gini", "Municipios_Name", "geometry"], axis=1).columns.tolist()  # Independent variables names
)
```

```python
print(model_4.summary)
```
